```python
#abre el archivo de entrada
with open("anotaciones/downreg_40vs30.blastx.txt", "r") as archivo:
    lineas = archivo.readlines()
```

```python
codigos_de_acceso = []
```

```python
for linea in lineas:
    partes = linea.split("|")
    if len(partes) >= 2:
        codigo_acceso = partes[1] 
        codigos_de_acceso.append(codigo_acceso)
```

```python
for codigo in codigos_de_acceso:
    print(codigo)
```

```python
with open("codacceso_downreg40vs30.txt", "w") as archivo_salida: 
    for codigo in codigos_de_acceso:
        archivo_salida.write(codigo + "\n")
    
```
